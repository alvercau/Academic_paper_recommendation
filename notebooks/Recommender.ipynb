# The recommender system

Takes a title, and returns the 3 most similar papers based on the topics space of the papers that were taken into account.  
Warning: title has to be inputted in exactly the same format as it is in the db.

## Merging different topic spaces

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.externals import joblib
```

```python
df = joblib.load('topic_space_papers')
```

```python
df.head()
```

```python
df0 = joblib.load('topic_space_topic0')
df3 = joblib.load('topic_space_topic3')
```

```python
df_all = df.copy()
```

```python
df_all['topic01'] = 0
df_all['topic02'] = 0
df_all['topic03'] = 0
df_all['topic04'] = 0
df_all['topic05'] = 0
df_all['topic06'] = 0
df_all['topic31'] = 0
df_all['topic32'] = 0
df_all['topic33'] = 0
df_all['topic34'] = 0
df_all['topic35'] = 0
```

```python
df_all.head()
```

```python
df_all.loc[:,'topic01']
```

```python
for i in range(len(df0)):
    title = df0.title[i]
    index = df_all[df_all.title == title].index
    df_all.loc[index, 'topic01'] = df_all.loc[index, 'topic0'] * df0.loc[i].topic0
    df_all.loc[index, 'topic02'] = df_all.loc[index, 'topic0'] * df0.loc[i].topic1
    df_all.loc[index, 'topic03'] = df_all.loc[index, 'topic0'] * df0.loc[i].topic2
    df_all.loc[index, 'topic04'] = df_all.loc[index, 'topic0'] * df0.loc[i].topic3
    df_all.loc[index, 'topic05'] = df_all.loc[index, 'topic0'] * df0.loc[i].topic4
    df_all.loc[index, 'topic06'] = df_all.loc[index, 'topic0'] * df0.loc[i].topic5
```

```python
for i in range(len(df3)):
    title = df3.title[i]
    index = df_all[df_all.title == title].index
    df_all.loc[index, 'topic31'] = df_all.loc[index, 'topic0'] * df3.loc[i].topic0
    df_all.loc[index, 'topic32'] = df_all.loc[index, 'topic0'] * df3.loc[i].topic1
    df_all.loc[index, 'topic33'] = df_all.loc[index, 'topic0'] * df3.loc[i].topic2
    df_all.loc[index, 'topic34'] = df_all.loc[index, 'topic0'] * df3.loc[i].topic3
    df_all.loc[index, 'topic35'] = df_all.loc[index, 'topic0'] * df3.loc[i].topic4
```

```python
df_all.tail(10)
```

## Calculating similarities
### Topic similarities

```python
cosine_similarity(df_all.iloc[:, 1:]).shape
```

```python
def calculate_paper_similarities(df):
    """takes df with titles and their topic space and returns the distances between the papers"""
    similarities = cosine_similarity(df.iloc[:, 1:])
    out = pd.DataFrame(similarities, index=df.title, columns=df.title)
    return out
```

```python
calculate_paper_similarities(df_all)
```

```python
similarities = calculate_paper_similarities(df_all)
```

```python
top_sim = similarities.as_matrix()
```

### Document similarities

```python
dtm_count = joblib.load('countvector.pkl')
```

```python
docs_sim = cosine_similarity(dtm_count)
```

```python
df_docs_sim = pd.DataFrame(docs_sim, index=list(similarities), columns=list(similarities))
```

```python
df_docs_sim
```

### Keyword similarities

```python
keyword_count = joblib.load('countvector_keywords.pkl')
key_sim = cosine_similarity(keyword_count)
```

```python
df_key_sim = pd.DataFrame(key_sim, index=list(similarities), columns=list(similarities))
```

```python
df_key_sim.tail()
```

## Recommender

```python
def weigh_scores(alpha, beta, theta):
    """parameters indicate the weight to assign to the topic, document and keyword similarity respectively"""
    weighted = alpha*top_sim + beta*docs_sim + theta*key_sim
    return  pd.DataFrame(weighted, index=list(similarities), columns=list(similarities))
```

```python
weighted = weigh_scores(2, 1, 3)
```

```python
def recommend(title):
    scores = list(weighted.loc[title])
    maxi = sorted(scores, reverse=True)[1:4]
    indexes = [scores.index(maxi[i]) for i in range(len(maxi))]
    print('You might want to look into: \n')
    for i in indexes:
        print(list(weighted)[i])
    #return [list(weighted)[i] for i in indexes]
```

```python
recommend('The feature structure of pronouns: a probe into multidimensional paradigms')
```

```python
all_recommendations = []

for title in df.title:
    all_recommendations.append(recommend(title))
    
```

```python
df_all_recommendations = pd.DataFrame(all_recommendations)
```

```python
df_all_recommendations.head()
```

```python
df_all_recommendations['paper'] = list(df.title)
```

```python
df_all_recommendations.rename(columns={0: 'topic0', 1: 'topic1', 2: 'topic2'}, inplace=True)
```
